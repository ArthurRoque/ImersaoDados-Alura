## An√°lise inicial da tabela de experimentos

```python
import pandas as pd
url = "https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true"
experimentos = pd.read_csv(url, compression='zip')
experimentos
```

Antes de partir para a an√°lise, seria interessante filtrar os elementos e dividi-los por tratamento. 

```python
experimentos_com_droga = experimentos[experimentos["tratamento"] == "com_droga"]
experimentos_com_controle = experimentos[experimentos["tratamento"] == "com_controle"]
```

### Desafio 1 - Por que a classe com controle √© t√£o pequena se comparada com a classe com droga?

O interesse nos experimentos com controle estava apenas na droga 'cacb2b860'. Pois s√≥ ela foi utilizada nesse tipo de experimento.

```python
experimentos_com_controle["droga"].value_counts(normalize=True)
```

### Desafio 6 - Renomear as colunas removendo o h√≠fen.

```python
# O desafio 6 foi feito antes do desafio 5 pois √© preciso renomear as colunar para facilitar a query.

#JEITO 1
colunas = {}
for coluna in experimentos.columns:
    colunas[coluna] = coluna.replace("-", "")
experimentos.rename(columns=colunas)

#JEITO 2 
experimentos.columns = experimentos.columns.str.replace("-", "")

experimentos
```

### Desafio 5 - Procurar na documenta√ß√£o o m√©todo query()

O m√©todo query √© um m√©todo para filtrar as filas de um dataframe utilizando o par√¢metro passado. 

```python
experimentos.query('droga == "87d714366"')
```

```python
experimentos.query('g0 > 8')
```

### Desafio 7 - Plotagem de gr√°ficos Matplotlib

```python
import matplotlib.pyplot as plt
```

* Gr√°fico aplitude g0 e g1 

```python
ranges = [f'g0 (de {experimentos["g0"].min()} a {experimentos["g0"].max()})', f'g1 (de {experimentos["g1"].min()} a {experimentos["g1"].max()})']
amplitude = [(experimentos["g0"].max() - experimentos["g0"].min()), (experimentos["g1"].max()- experimentos["g1"].min())]

plt.bar(ranges, amplitude, color="blue")

plt.title('Amplitude dos elementos g')
plt.show()
```
