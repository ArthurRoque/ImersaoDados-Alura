# Estatística, visualização de dados e distribuições

```python
import pandas as pd
url = "https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true"
experimentos = pd.read_csv(url, compression='zip')
experimentos
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

### Renomeando tabela

```python
mapa = {'droga':'composto'}
for coluna in experimentos.columns:
    if coluna != 'droga':
        mapa[coluna] = coluna.replace("-", "")
experimentos.rename(columns=mapa, inplace=True)
experimentos
```

### Gráfico dos cinco compostos mais utilizados 

```python
compostos_mais_utilizados = experimentos['composto'].value_counts().index[:5]
compostos_mais_utilizados
```

```python
experimento_compostos_mais_utilizados = experimentos.query('composto in @compostos_mais_utilizados')
experimento_compostos_mais_utilizados
```

```python
plt.figure(figsize=(8.5, 6))
ax = sns.countplot(x='composto', data=experimento_compostos_mais_utilizados)
ax.set_title('COMPOSTOS MAIS UTILIZADOS NO EXPERIMENTO')
plt.show()
```

## Desafio 1 e 2 - Ordenar o gráfico countplot e melhorar a vizualização

```python
plt.figure(figsize=(8.5, 6))
plt.rcParams["font.family"] = "serif"
ax = sns.countplot(x='composto', data=experimento_compostos_mais_utilizados, order=experimento_compostos_mais_utilizados['composto'].value_counts().index)

ax.set_title('OS 5 COMPOSTOS MAIS UTILIZADOS NO EXPERIMENTO', fontsize = 15)
ax.set_ylabel('QUANTIDADE')
ax.set_xlabel('COMPOSTO')
plt.xticks(rotation=35)
plt.show()
```

### Histograma 

O histograma é uma forma de representar a distribuições de dados agrupados. Quando é possível criar classes para uma variável, é comum o uso de histogramas para a visualização da distribuição dos elementos em cada classe. Enquanto no gráfico de pizza, relaciona-se duas variáveis, no histograma, relaciona-se a distribuição das frequências de uma única variável quantitativa.

* Classe: Na estatística, as classes são intervalos de uma variável.

Exemplo: 

Variável: Idade ([11,11,12,13,14,14,15,15,16,16,16,16,17,17,19])

classes: ([10 a 13], [13 a 16], [17 a 20])

```python
plt.figure(figsize=(20, 5.5))
experimentos['g0'].hist(bins=200, color='black')
plt.title('DISTRIBUIÇÃO DOS ELEMENTOS G0')
plt.show()
```

No exemplo a cima, nota-se que os elementos estão mais concentrados nos intervalos próximos de 0. Nota-se também que existem mais intervalos de elementos positivos do que de elementos negativos.

## Desafio 3 e 4 - Plotar histogramas com o seaborn e estudar o método .describe()

    O método DataFrame.describe() é um método da biblioteca pandas que retorna algumas estatísticas do dataframe. São elas: A quantidade(count), a média(mean), desvio padrão (std), mínima(min), o primeiro quartil(25%), a mediana ou segundo quatil(50%), terceiro quartil(75%) e a máxima(max).

### Extraindo as medidas estatísticas dos genes

```python
genes = experimentos.loc[:,'g0':'g771'] #Selecionando todas as linhas (:) e todas as colunas de genes ('g0':'g771') do experimento.
genes = genes.describe().T #Transpondo a tabela para que as estatísticas se tornem colunas e os genes linhas.
genes
```

### Plotando um histograma para cada medida estatística por meio de subplots

```python
fig, ax = plt.subplots(3,3, figsize = (20,10), constrained_layout = True) 
# fig: variável que armazena as características da figura total (figura com tamanho 720x576)
# ax: variável que armazena os subplots.

fig.suptitle('MEDIDAS ESTATÍSTICAS DAS CATEGORIAS DE GENES', fontsize=30)
fig.delaxes(ax[2,1])
fig.delaxes(ax[2,2])
# Não há a necessidadde de montar um histograma para a medida quantidade(count) uma vez que todas as categorias(g0,g1,g2,...) possuem a mesma quantidade de genes.
histograma_media = sns.histplot(data = genes, x = 'mean', bins = 100, ax=ax[0,0])
histograma_media.set(xlabel='MÉDIA', ylabel='')

histograma_desvio_padrao = sns.histplot(data = genes, x = 'std', bins = 100,ax=ax[0,1])
histograma_desvio_padrao.set(xlabel='DESVIO PADRÃO', ylabel='')

histograma_minima = sns.histplot(data = genes, x = 'min', bins = 100,ax=ax[0,2])
histograma_minima.set(xlabel='MÍNIMA', ylabel='')

histograma_primeiro_quartil = sns.histplot(data = genes, x = '25%', bins = 100,ax=ax[1,0])
histograma_primeiro_quartil.set(xlabel='PRIMEIRO QUARTIL', ylabel='')

histograma_mediana = sns.histplot(data = genes, x = '50%', bins = 100, ax=ax [1,1])
histograma_mediana.set(xlabel='MEDIANA', ylabel='')

histograma_terceiro_quartil = sns.histplot(data = genes, x = '75%', bins = 100, ax=ax [1,2])
histograma_terceiro_quartil.set(xlabel='TERCEIRO QUARTIL', ylabel='')

histograma_maxima = sns.histplot(data = genes, x = 'max', bins = 100,ax=ax[2,0])
histograma_maxima.set(xlabel='MÁXIMA', ylabel='')

plt.show()
```

### Boxplot

O Boxplot, também conhecido como diagrama de caixas, é uma ferramenta que permite a visualização da distribuição dos dados, assim como o histograma. O boxplot diferencia-se do histograma pois marca medidas estatísticas (média, máxima, mínima e quartis) além de proporcionar uma melhor vizualização dos outliers.

<img src='https://i0.wp.com/www.abgconsultoria.com.br/blog/wp-content/uploads/boxplot-1-1024x664.png?resize=1024%2C664'>

**Fonte da imagem:** https://operdata.com.br/blog/como-interpretar-um-boxplot

```python
sns.boxplot(data=experimentos, x='g0', y='dose')
plt.show()
```

```python
sns.boxplot(data=experimentos, x='g0', y='tratamento')
plt.show()
```

## Desafio 5 - Refletir sobre  o tamanho das  visualizações 

O tamanho das plotagens não só define se uma visualização é boa ou ruim, como também pode proporcionar uma maior riqueza nos  gráficos. Se tomarmos o histograma da distribuição do gene 0 como exemplo e tentarmos montá-lo com 290 bins sem ajustar o tamanho, este é o resultado:

```python
experimentos['g0'].hist(bins=290)
plt.show()
```

Por outro lado, se ajustarmos o tamanho do eixo x da figura:

```python
plt.figure(figsize=(20, 5.5))
experimentos['g0'].hist(bins=400)
plt.show()
```

Concluí-se portanto que um gráfico ajustado da maneira correta permite que uma maior quantidade de riqueza seja inserida. O mesmo vale para os histogramas e todas as outras ferramentas gráficas.
